# Symbolic Regression with gplearn


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from gplearn.genetic import SymbolicClassifier
from collections import Counter

# Load and label Train files
train_norm = pd.read_csv("normal_train2.csv").assign(Target=0).dropna()
train_bact = pd.read_csv("bacteria_train2.csv").assign(Target=1).dropna()
train_viru = pd.read_csv("virus_train2.csv").assign(Target=1).dropna()

# Load and label Test files
test_norm = pd.read_csv("normal_test2.csv").assign(Target=0).dropna()
test_bact = pd.read_csv("bacteria_test2.csv").assign(Target=1).dropna()
test_viru = pd.read_csv("virus_test2.csv").assign(Target=1).dropna()

# Combine into Binary Sets
train_df = pd.concat([train_norm, train_bact, train_viru], ignore_index=True)
test_df = pd.concat([test_norm, test_bact, test_viru], ignore_index=True)

# Define feature list
features = ['EntropyMean', 'EntropyMax', 'Entropy90', 'Haze', 'Solidity', 
           'Contrast', 'Homogeneity', 'Skewness', 'Kurtosis', 'FractalDim']


X_train, y_train = train_df[features], train_df['Target']
X_test, y_test = test_df[features], test_df['Target']
```

```python
# APPLY CONSTRAINED SYMBOLIC REGRESSION/CLASSIFICATION

sc = SymbolicClassifier(
    population_size=5000, 
    generations=50, 
    tournament_size=20,
    parsimony_coefficient=0.005,
    function_set=('add', 'sub', 'mul', 'div', 'log', 'sqrt'),
    max_samples=0.9,
    p_crossover=0.7,
    p_subtree_mutation=0.1,
    p_hoist_mutation=0.1,
    p_point_mutation=0.1,
    stopping_criteria=0.0,
    verbose=1,
    random_state=42
)

sc.fit(X_train, y_train)

print("\n--- Best Formula ---")
print(sc._program)
print(f"Binary Test Accuracy: {sc.score(X_test, y_test):.4f}")
```
